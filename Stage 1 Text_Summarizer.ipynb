```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = "https://en.wikipedia.org/wiki/Alexander_the_Great"
```

```python
page = urlopen(url)
html = page.read().decode("utf-8")
soup = BeautifulSoup(html,"html.parser")
```

```python
text = str(soup)
text
```
